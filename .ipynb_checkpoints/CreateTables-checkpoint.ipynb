```python
def dataType(val, current_type):
    try:
        # Evaluates numbers to an appropriate type, and strings an error
        t = ast.literal_eval(val)
        #print(type(t))
    except ValueError:
       # print("val error")
        return 'varchar'
    except SyntaxError:
       # print('syntax error')
        return 'varchar'
    if type(t) in [int, float]:
        #return 'varchar'
#         if (type(t) in [int]) and current_type not in ['float', 'varchar']:
#            # Use smallest possible int type
#             if (-32768 < t < 32767) and current_type not in ['int', 'bigint']:
#                 return 'varchar'#smallint
#             elif (-2147483648 < t < 2147483647) and current_type not in ['bigint']:
#                 return 'varchar'#int
#             else:
#                 return 'bigint'

       # print( type(t), current_type)
        if type(t) is int and current_type not in ['varchar']:
           # print('decimal')
            return 'decimal'
        else: 
           # print('else varchar')
            return'varchar'
    else:
       # print('last varchar')
        return 'varchar'
```

```python
def createTable(recentFile, ending):
    
    f = open(recentFile, 'r')#.read().split('\n')
    reader = csv.reader(f)
    longest, headers, type_list = [], [], []
  #  headers = features
  #  for x in headers:
    #    longest.append(0)
     #   type_list.append('')
    
    firstLine = True
    for row in reader:
        if firstLine:
            firstLine = False
        if len(headers) == 0:
            headers = row
            for col in row:
                longest.append(0)
                type_list.append('')
        else:
            for i in range(len(row)):
            # NA is the csv null value
                if type_list[i] == 'varchar' or row[i] == 'NA':
                    pass
                else:
                    var_type = dataType(row[i], type_list[i])
                    #print(var_type)
                    
                    type_list[i] = var_type
            if len(row[i]) > longest[i]:
                longest[i] = len(row[i])
        #f.close()

    statement = 'create table {} ( \n'.format(ending)

    for i in range(len(headers)):
        if type_list[i] == 'varchar':
            statement = (statement + '{} varchar(255), \n').format(headers[i].upper())
            
        else:
            statement = (statement + '{} {}' + ', \n').format(headers[i].upper(), type_list[i])
    print(statement)
    return statement[:-3] + ');'
    
```

```python
from os import listdir
from os.path import isfile, join
import os
import csv
cuDirectory = "C:/Users/Owner/Documents/PRACTICUM PROJECTS/Arkatechture/FDIC_S3/FDIC_S3"

features = list()
tmp =0
lastName = ""
#Get all sub directories 
tablesFeatures = []
tablesTypes = []
for x in os.walk(cuDirectory):
    for file in x[2]:
        name = file[0:5]
        #Don't need to repeat a file
        if name == lastName:
            continue
        #Grab all features
        else: 
            features = []
            typeList = []
            with open(cuDirectory+"/"+file, 'r') as f:
                d_reader = csv.DictReader(f)
                #get fieldnames from DictReader object and store in list
                headers = d_reader.fieldnames
                for h in headers:
                    features.append(h)
                
                #Look at the first row for each header
                firstLine = True
                for row in d_reader:
                    if not firstLine:
                        for col in row:
                            typeList.append(getType(row[col]))
                    if not firstLine:
                        break
                    firstLine = False
                    
                
                
            tablesFeatures.append(features)
            tablesTypes.append(typeList)
        lastName = name
```

```python
def getType(val):
    try:
        # Evaluates numbers to an appropriate type, and strings an error
        t = ast.literal_eval(val)
        #print(type(t))
    except ValueError:
       # print("val error")
        return 'varchar'
    except SyntaxError:
       # print('syntax error')
        return 'varchar'
    if type(t) in [int, float]:
        return 'decimal'
    else:
        return 'varchar'
```

```python
createTales = []
for i in range(0,len(tablesFeatures)):
    statement = 'create table {} ( \n'
    for j in range(0,len(tablesFeatures[i])):
        varType = tablesTypes[i][j]
        featureName = tablesFeatures[i][j]
        
        
        if varType == 'varchar':
            statement = (statement + featureName.upper() +  ' varchar(255), \n')
        else:
            statement = (statement + featureName.upper() + " " + varType + ', \n')
    createTales.append(statement)
```
